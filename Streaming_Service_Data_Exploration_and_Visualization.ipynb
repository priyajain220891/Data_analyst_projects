**Task 5- explode the nested value  in a column**

 **explode the nested value in director column**

```python
director=df['director'].apply(lambda x: str(x).split(', ')).tolist()
df_director=pd.DataFrame(director,index=df['title'])
df_director=df_director.stack()
df_director=pd.DataFrame(df_director)
df_director.reset_index(inplace=True)
df_director=df_director[['title',0]]
df_director.columns=['title','director']
df_director
```

 **explode the nested value in cast column**

```python
cast=df['cast'].apply(lambda x: str(x).split(', ')).tolist()
df_cast=pd.DataFrame(cast,index=df['title'])
df_cast=df_cast.stack()
df_cast=pd.DataFrame(df_cast)
df_cast.reset_index(inplace=True)
df_cast=df_cast[['title',0]]
df_cast.columns=['title','cast']
df_cast
```

 **explode the nested value in country column**

```python
country=df['country'].apply(lambda x: str(x).split(', ')).tolist()
df_country=pd.DataFrame(country,index=df['title'])
df_country=df_country.stack()
df_country=pd.DataFrame(df_country)
df_country.reset_index(inplace=True)
df_country=df_country[['title',0]]
df_country.columns=['title','country']
df_country
```

**explode the nested value in listed_in column**

```python
genres=df['listed_in'].apply(lambda x: str(x).split(', ')).tolist()
df_genres=pd.DataFrame(genres,index=df['title'])
df_genres=df_genres.stack()
df_genres=pd.DataFrame(df_genres)
df_genres.reset_index(inplace=True)
df_genres=df_genres[['title',0]]
df_genres.columns=['title','genres']
df_genres
```

**best time to launch a TV show or movies**

```python
df_final['date_added'] = pd.to_datetime(df_final['date_added'])
df_final['week_added'] = df_final['date_added'].dt.isocalendar().week
df_final['month_added'] =df_final['date_added'].dt.month
```

```python
# Group by week and count TV shows and movies separately
tv_shows_by_week = df_final[df_final['type'] == 'TV Show'].groupby('week_added')['type'].count()
movies_by_week = df_final[df_final['type'] == 'Movie'].groupby('week_added')['type'].count()

# Group by month and count TV shows and movies separately
tv_shows_by_month = df_final[df_final['type'] == 'TV Show'].groupby('month_added')['type'].count()
movies_by_month =df_final[df_final['type'] == 'Movie'].groupby('month_added')['type'].count()
```

```python


# Plot for best week to release TV shows and movies
plt.figure(figsize=(10, 5))
plt.plot(tv_shows_by_week.index, tv_shows_by_week.values, label='TV Shows')
plt.plot(movies_by_week.index, movies_by_week.values, label='Movies')
plt.xlabel('Week of the Year')
plt.ylabel('Count')
plt.title('Best Week to Release TV Shows and Movies')
plt.legend()
plt.show()

# Plot for best month to release TV shows and movies
plt.figure(figsize=(10, 5))
plt.plot(tv_shows_by_month.index, tv_shows_by_month.values, label='TV Shows')
plt.plot(movies_by_month.index, movies_by_month.values, label='Movies')
plt.xlabel('Month')
plt.ylabel('Count')
plt.title('Best Month to Release TV Shows and Movies')
plt.legend()
plt.show()
```

 **Find After how many days the movie will be added to the platform after the release of
the movie.**.

```python
# Convert 'date_added' and 'release_year' columns to datetime type
df_final['date_added'] = pd.to_datetime(df_final['date_added'])
df_final['release_date'] = pd.to_datetime(df_final['release_year'], format='%Y')

# Calculate the difference in days
df_final['days_to_add'] = (df_final['date_added'] - df_final['release_date']).dt.days

# Filter only movies
movies_data = df_final[df_final['type'] == 'Movie']

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(movies_data['days_to_add'], bins=30, edgecolor='black', alpha=0.7)
plt.xlabel('Days to Add After Release')
plt.ylabel('Frequency')
plt.title('Distribution of Days to Add for Movies on Netflix')
plt.show()
```
